```python
import pandas as pd
import hydrological_signatures
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
# Colocar dataframe no modelo correto para a aplicar ccrs
df_assinaturas = hydrological_signatures.result()
df_assinaturas = df_assinaturas.T
df_assinaturas = df_assinaturas.reset_index(drop=True)
```

```python
# Obter a matriz de correlação
my_r = df_assinaturas.corr(method='spearman', numeric_only=False)
my_r
```

```python
def correlation_matrix(data):
    import seaborn as sns
    sns.set(rc={"figure.dpi":300, 'savefig.dpi':300})
    sns.set_context('notebook')
    sns.set_style("ticks")
    
    corr = data.corr(method='spearman', numeric_only=False)
    
    ax = sns.heatmap(corr,
        vmin=-1, vmax=1, center=0,
        cmap=sns.diverging_palette(20, 220, n=200),
        square=True
    )
    ax.set_xticklabels(
        ax.get_xticklabels(),
        rotation=45,
        fontsize=7,
        horizontalalignment='right'
    )
    ax.set_yticklabels(
        ax.get_yticklabels(),
        fontsize=7,
    )
    
    return
```

```python
correlation_matrix(df_assinaturas)
```
