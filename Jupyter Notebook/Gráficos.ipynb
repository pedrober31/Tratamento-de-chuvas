# Gráfico de Tira

```python
from Cluster_KMeans import estacoes_cluster
import matplotlib.pyplot as plt
import hydrological_signatures
import seaborn
```

```python
df_assinaturas = hydrological_signatures.result()
df_assinaturas = df_assinaturas.T
```

```python
df_k = estacoes_cluster(2, df_assinaturas)
```

```python
colunas = df_assinaturas.columns
colunas = ' '.join(colunas)
colunas = colunas.upper()
colunas = colunas.split(' ')
```

```python
df_assinaturas.columns = colunas
df_assinaturas['Cluster'] = df_k['Labels']
```

```python
df_assinaturas['Cluster'] = df_assinaturas['Cluster'] + 1
```

```python
df_assinaturas
```

```python
fig, axes = plt.subplots(nrows=2, ncols=4, sharex='col', figsize=(25, 10))

seaborn.set(style='whitegrid')

seaborn.stripplot(data=df_assinaturas, x='Cluster', y='QSP', hue='Cluster', palette='tab10', size=7.5, linewidth=1, edgecolor='k', ax=axes[0, 0], jitter=True, legend=False)
seaborn.stripplot(data=df_assinaturas, x='Cluster', y='CVQ', hue='Cluster', palette='tab10', size=7.5, linewidth=1, edgecolor='k', ax=axes[0, 1], jitter=True, legend=False)
seaborn.stripplot(data=df_assinaturas, x='Cluster', y='BFI', hue='Cluster', palette='tab10', size=7.5, linewidth=1, edgecolor='k', ax=axes[0, 2], jitter=True, legend=False)
seaborn.stripplot(data=df_assinaturas, x='Cluster', y='Q5', hue='Cluster', palette='tab10', size=7.5, linewidth=1, edgecolor='k', ax=axes[0, 3], jitter=True, legend=False)
seaborn.stripplot(data=df_assinaturas, x='Cluster', y='HFD', hue='Cluster', palette='tab10', size=7.5, linewidth=1, edgecolor='k', ax=axes[1, 0], jitter=True, legend=False)
seaborn.stripplot(data=df_assinaturas, x='Cluster', y='Q95', hue='Cluster', palette='tab10', size=7.5, linewidth=1, edgecolor='k', ax=axes[1, 1], jitter=True, legend=False)
seaborn.stripplot(data=df_assinaturas, x='Cluster', y='LOWFR', hue='Cluster', palette='tab10', size=7.5, linewidth=1, edgecolor='k', ax=axes[1, 2], jitter=True, legend=False)
seaborn.stripplot(data=df_assinaturas, x='Cluster', y='HIGHFRVAR', hue='Cluster', palette='tab10', size=7.5, linewidth=1, edgecolor='k', ax=axes[1, 3], jitter=True, legend=False)

plt.suptitle('Stripplot', fontsize='xx-large', weight='bold')

# axes[0, 0].legend(fontsize='large', markerscale=1.5)
# axes[0, 1].legend(fontsize='large', markerscale=1.5)
# axes[0, 2].legend(fontsize='large', markerscale=1.5)
# axes[0, 3].legend(fontsize='large', markerscale=1.5)
# axes[1, 0].legend(fontsize='large', markerscale=1.5)
# axes[1, 1].legend(fontsize='large', markerscale=1.5)
# axes[1, 2].legend(fontsize='large', markerscale=1.5)
# axes[1, 3].legend(fontsize='large', markerscale=1.5)

axes[0, 0].set_ylabel('QSP', fontsize='large', weight='bold')
axes[0, 1].set_ylabel('CVQ', fontsize='large', weight='bold')
axes[0, 2].set_ylabel('BFI', fontsize='large', weight='bold')
axes[0, 3].set_ylabel('Q5', fontsize='large', weight='bold')
axes[1, 0].set_xlabel('Cluster', fontsize='large', weight='bold')
axes[1, 0].set_ylabel('HFD', fontsize='large', weight='bold')
axes[1, 1].set_xlabel('Cluster', fontsize='large', weight='bold')
axes[1, 1].set_ylabel('Q95', fontsize='large', weight='bold')
axes[1, 2].set_xlabel('Cluster', fontsize='large', weight='bold')
axes[1, 2].set_ylabel('LOWFR', fontsize='large', weight='bold')
axes[1, 3].set_xlabel('Cluster', fontsize='large', weight='bold')
axes[1, 3].set_ylabel('HIGHFRVAR', fontsize='large', weight='bold')

plt.show()
```
